Using t-test to check for changes in growth of GDP in Iceland pre and post 2008 crisis.

---



```python
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('iceland_gdp_quarterly.xlsx')
df
```

```python
pre_crisis = df[df['Period'] <= '2007-Q4']['GDP_growth']
crisis     = df[(df['Period'] >= '2008-Q1') & (df['Period'] <= '2010-Q4')]['GDP_growth']
recovery   = df[df['Period'] >= '2011-Q1']['GDP_growth']
print(f"Pre-crisis mean:  {pre_crisis.mean():.2f}%  (n={len(pre_crisis)})")
print(f"Crisis mean:      {crisis.mean():.2f}%  (n={len(crisis)})")
print(f"Recovery mean:    {recovery.mean():.2f}%  (n={len(recovery)})")
```

```python
t1, p1 = stats.ttest_ind(pre_crisis, recovery)
```

```python
print(f"\nPre-crisis vs Recovery: t={t1:.4f}, p={p1:.4f}")
```
